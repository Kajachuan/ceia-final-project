```python
import rasterio
import numpy as np
import torch

from PIL import Image
from google.colab import drive
from torchvision.transforms import v2
```

```python
drive.mount('/content/drive')
```

```python
full_image = 'drive/MyDrive/CEIA/Trabajo Final/Dataset/T20HPJ_20250115T135701_TCI_10m.jp2'
```

```python
dataset_pil = Image.open(full_image)
```

```python
init_transform = v2.Compose([
    v2.ToImage(),
    v2.ToDtype(torch.float, scale=True)
])
```

```python
dataset_tensor = init_transform(dataset_pil)
```
